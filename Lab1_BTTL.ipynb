Dữ liệu rỗng rất ít và chiếm phần trăm rất thấp

Tiền xử lý dữ liệu

```python
bank['contact'].describe()
```

```python
bank=bank.drop(['contact'],axis=1)
```

```python
bank.head(12)
```

```python
bank['duration'].isnull().sum()
```

```python
for dataset in [bank]:
    mean = bank["duration"].mean() 
    std = bank["duration"].std() 
    is_null = dataset["duration"].isnull().sum()
    rand_duration = np.random.randint(mean-std,mean+std,size=is_null)
    duration_slice  = dataset["duration"].copy()
    duration_slice[np.isnan(duration_slice)]= rand_duration
    dataset["duration"]=duration_slice
    dataset["duration"]=bank["duration"].astype(int)
bank["duration"].isnull().sum() 
```

```python
bank['default'].isnull().sum()
```

```python
cothetindung ={"no":0,"yes":1}

for dataset in [bank]:
    dataset['default']=dataset['default'].map(cothetindung)
```

```python
bank.head(12)
```

```python
#Tương tự vậy biểu diễn housing, loan
yesno ={"no":0,"yes":1}

for dataset in [bank]:
    dataset['housing']=dataset['housing'].map(yesno)
    dataset['loan']=dataset['loan'].map(yesno)
```

```python
for dataset in [bank]:
    dataset['age']=dataset['age'].astype(int)
    dataset.loc[dataset['age']<=11,'age']=0
    dataset.loc[(dataset['age']>11) & (dataset['age']<=18),'age']=1
    dataset.loc[(dataset['age']>18) & (dataset['age']<=22),'age']=2
    dataset.loc[(dataset['age']>22) & (dataset['age']<=27),'age']=3
    dataset.loc[(dataset['age']>27) & (dataset['age']<=33),'age']=4
    dataset.loc[(dataset['age']>33) & (dataset['age']<=40),'age']=5
    dataset.loc[(dataset['age']>40) & (dataset['age']<=66),'age']=6
    dataset.loc[dataset['age']>66,'age']=7
```
